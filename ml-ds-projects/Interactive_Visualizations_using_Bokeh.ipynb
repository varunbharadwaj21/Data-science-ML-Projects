```python
from bokeh.transform import factor_cmap, factor_mark
BODY_STYLE = ['sedan', 'hatchback', 'Other']
MARKERS = ['hex', 'circle_x', 'triangle']

p = figure(title = "Automobile Dataset")
p.xaxis.axis_label = 'Engine Size'
p.yaxis.axis_label = 'Wheel Base'

p.scatter("engine-size", "wheel-base", source=auto_df, legend_field="body-style", fill_alpha=0.4, size=12,
          marker=factor_mark('body-style', MARKERS, BODY_STYLE),
          color=factor_cmap('body-style', 'Category10_3', BODY_STYLE))

show(p)
```

# grid plot

```python
from bokeh.models import CDSView, IndexFilter
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
from bokeh.layouts import gridplot
from bokeh.io import show


# Use 'filter' instead of 'filters' and pass the filter object directly
view = CDSView(filter=IndexFilter([70, 90, 110,130]))

source = ColumnDataSource(auto_df)

tools = ["box_select", "hover", "reset"]
# Corrected argument names from plot_height to height and plot_width to width
p = figure(height=300, width=300, tools=tools)
p.circle(x="engine-size", y="wheel-base", size=10, hover_color="red", source=source)

# Corrected argument names from plot_height to height and plot_width to width
p_filtered = figure(height=300, width=300, tools=tools)
# The source is provided here when using the view
p_filtered.circle(x="engine-size", y="wheel-base", size=10, hover_color="red", source=source, view=view)

show(gridplot([[p, p_filtered]]))
```

# other grid plot

```python
from bokeh.models import BooleanFilter
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
from bokeh.layouts import gridplot
from bokeh.io import show

# Create a list of booleans based on the 'wheel-base' column
booleans = [True if y_val > 110 else False for y_val in source.data['wheel-base']]

# Create a BooleanFilter instance with the boolean list
boolean_filter = BooleanFilter(booleans)

# Create the CDSView with the single BooleanFilter instance
view = CDSView(filter=boolean_filter)

tools = ["box_select", "hover", "reset"]
# Corrected argument names from plot_height to height and plot_width to width
p = figure(height=300, width=300, tools=tools)
p.circle(x="engine-size", y="wheel-base", size=10, hover_color="red", source=source)

# Corrected argument names from plot_height to height and plot_width to width
p_filtered = figure(height=300, width=300, tools=tools,
                    x_range=p.x_range, y_range=p.y_range)
# The source is provided here when using the view, which is correct
p_filtered.circle(x="engine-size", y="wheel-base", size=10, hover_color="red", source=source, view=view)

show(gridplot([[p, p_filtered]]))
```

```python
#GroupFilter

from bokeh.models import GroupFilter, CDSView
from bokeh.plotting import figure
from bokeh.layouts import gridplot
from bokeh.io import show
from bokeh.models import ColumnDataSource # Import ColumnDataSource

# Assuming 'source' is already defined from a previous cell, if not, define it here:
# source = ColumnDataSource(auto_df)

# Remove the source argument from the CDSView initialization
# Corrected attribute name from filters to filter
view1 = CDSView(filter=GroupFilter(column_name='body-style', group='hatchback'))
# Corrected the keys in the dictionary to use 'height' and 'width'
plot_size_and_tools = {'height': 300, 'width': 300,
                        'tools':['box_select', 'reset', 'help']}

p1 = figure(title="Full data set", **plot_size_and_tools)
# Explicitly set the radius for the circle glyph
p1.circle(x='engine-size', y='wheel-base', source=source, color='black', radius=5)

p2 = figure(title="Hatchback only", x_range=p1.x_range, y_range=p1.y_range, **plot_size_and_tools)
# Explicitly set the radius for the circle glyph
p2.circle(x='engine-size', y='wheel-base', source=source, view=view1, color='red', radius=5)

show(gridplot([[p1, p2]]))
```

# vertical bar graph

```python
from bokeh.palettes import Spectral6
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, HoverTool # Import necessary models

body_style = ['sedan', 'hatchback', 'wagon', 'convertible']
counts = [(auto_df['body-style'].values == 'sedan').sum(),(auto_df['body-style'].values == 'hatchback').sum(),(auto_df['body-style'].values == 'wagon').sum(),(auto_df['body-style'].values == 'convertible').sum()]
source = ColumnDataSource(data=dict(body_style=body_style, counts=counts, color=Spectral6))

# Use 'height' instead of 'plot_height'
p = figure(x_range=body_style, y_range=(0,200), height=250, title="Body Style of car Counts",
           toolbar_location=None, tools="")

p.vbar(x='body_style', top='counts', width=0.9, color='color', legend_field="body_style", source=source)

p.xgrid.grid_line_color = None
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

show(p)
```

# vertical bar graph of cylinders

```python
from bokeh.palettes import Spectral5

auto_df['body-style']= auto_df['body-style'].astype(str)
group = auto_df.groupby('body-style')

# Creating a ColumnDataSource directly from a groupby object is not standard Bokeh practice.
# It's better to create a DataFrame or a dictionary from the grouped data
# if you intend to use the aggregated information. However, looking at the vbar call,
# it seems you want to plot the distribution of 'engine-size' for each 'body-style'
# across the entire dataset, not the aggregated values from the groupby object.
# Therefore, we will use the original auto_df as the source for the glyph.
# The 'group' object is used for the x_range, which is correct.
# source = ColumnDataSource(group) # This line can be removed

cyl_cmap = factor_cmap('body-style', palette=Spectral5, factors=sorted(auto_df['body-style'].unique()))

# Change plot_height to height and plot_width to width
p = figure(height=350, x_range=group, title="Engine size by # Cylinders",
           toolbar_location=None, tools="")

# Pass auto_df directly as the source for the vbar glyph.
# This assumes you want to plot each individual data point's engine-size
# for each body style. If you intend to plot an aggregation (like mean or sum),
# you would need to calculate that aggregation from the 'group' object and
# create a ColumnDataSource from the aggregated data.
p.vbar(x='body-style', top='engine-size', width=1, source=auto_df,
       line_color=cyl_cmap, fill_color=cyl_cmap)

p.y_range.start = 0
p.xgrid.grid_line_color = None
p.xaxis.axis_label = "some stuff"
p.xaxis.major_label_orientation = 1.2
p.outline_line_color = None
show(p)
```

```python
from bokeh.sampledata.sprint import sprint
from bokeh.io import output_file, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource

output_file("sprint.html")

sprint.Year = sprint.Year.astype(str)
group = sprint.groupby('Year')
source = ColumnDataSource(group)

# Corrected parameter names from plot_width/plot_height to width/height
p = figure(y_range=group, x_range=(9.5,12.7), width=400, height=550, toolbar_location=None,
           title="Time Spreads for Sprint Medalists (by Year)")
p.hbar(y="Year", left='Time_min', right='Time_max', height=0.4, source=source)

p.ygrid.grid_line_color = None
p.xaxis.axis_label = "Time (seconds)"
p.outline_line_color = None

show(p)
```
